```python
class Node:
    def __init__(self, data):
        self.data = data
        self.Next = None
        
        
class CircularLinkedList:
    #constructor to creat empty linked list
    def __init__(self):
        self.head = None
        
    def add_in_begginning(self, new_data):
        new_node = Node(new_data)
        temp = self.head    
        new_node.next = self.head

        if self.head is not None:
             while(temp.next != self.head):
                temp = temp.next
                temp.next = new_node
        
                new_node.next = self.head
    def add_at_particular_position(self,new_data,position):
        new_node=Node(new_data)
        if self.head == None:
            self.head = new_node
            new_node.next = self.head
        else:
            temp=self.head
            count=1
            while(temp):
                if position == count:
                    temp_node=temp.next
                    new_node.next=temp_node
                    temp.next=new_node
                    break
                count+=1
                temp=temp.next
                
    def add_at_the_end(self,new_data):
        new_node = Node(new_data)
        
        if(self.head == None):
            self.head=new_node
         
        temp= self.head
        while(temp.next):
            temp=temp.next
           
        temp.next=new_node
    
    def print_linked_list(self):
        temp=self.head
        if self.head is not None:
            while(True):
                print(temp.data)
                temp=temp.next
                if temp == self.head:
                    print("same ")
                    break
```
