```python
import pandas as pd
import numpy as np
import csv
```

```python
def function_one():
    
    Energy = pd.read_excel('Energy+Indicators.xls', skiprows = 17, skipfooter = 38)

    Energy.rename(columns = {"Unnamed: 1" : "Country", "Petajoules" : "Energy Supply", "Gigajoules":"Energy Supply per Capita", "%":"%renewable"}, inplace = True)

    Energy.drop(columns = ['Unnamed: 0','Unnamed: 2'], axis = 1, inplace = True)

    Energy.replace("..." , np.nan, inplace = True)

    Energy["Energy Supply"] = 1000000 * Energy["Energy Supply"]

    Energy.replace('United States of America','United States', inplace = True)
    Energy.replace('United Kingdom of Great Britain and Northern Ireland','United Kingdom', inplace=True)
    Energy.replace('China, Hong Kong Special Administrative Region','Hong Kong', inplace=True)
    Energy.replace('Republic of Korea','South Korea', inplace=True)
    Energy.replace('Bolivia (Plurinational State of)','Bolivia', inplace=True)

    GDP = pd.read_csv('world_bank.csv', skiprows = 4)
    GDP.rename(columns = {'Country Name':'Country'}, inplace = True)

    GDP.drop(GDP.iloc[:,2:50], inplace = True, axis = 1)

    GDP.drop(columns = ['Country Code'], axis =1, inplace = True)

    GDP.replace('Korea, Rep.' , 'South Korea', inplace=True)
    GDP.replace('Iran, Islamic Rep.' , 'Iran', inplace=True)
    GDP.replace('Hong Kong SAR, China' , 'Hong Kong', inplace=True)

    ScimEn = pd.read_excel('scimagojr-3.xlsx')

    Ranks = ScimEn.head(15)

    merge1 = pd.merge(Ranks,Energy, on = 'Country', how = 'left')

    final = pd.merge(merge1,GDP, on = 'Country', how = 'left')

    final.set_index("Country", inplace = True)
    
    return final
```

```python
function_one()
```

```python
def function_two():
    return ((GDP.shape[0] - function_one().shape[0]) + (Energy.shape[0] - function_one().shape[0]))
```

```python
def function_three():
    rows = ['2006', '2007', '2008',
           '2009', '2010', '2011', '2012', '2013', '2014', '2015']
    AvgGDP = function_one()[rows].mean(axis=1)
    AvgGDP.sort_values(ascending = False, inplace = True)
    return AvgGDP
```

```python
function_three()
```

```python
def function_four():
    variable = function_one()
    variable['AvgGDP'] = function_three()
    variable.sort_values('AvgGDP', ascending=False,inplace =True)
    final = variable.iloc[5]['2015']
    initial = variable.iloc[5]['2006']
    return abs(final - initial)
```

```python
def function_seven():
    variable = function_one()
    variable['Self Citations to Total Citations'] = variable['Self-citations']/variable['Citations']
    return variable['Self Citations to Total Citations'].idxmax(),variable['Self Citations to Total Citations'].max()
```

```python
def function_eight():
    variable = function_one()
    variable['Population'] = variable['Energy Supply']/variable['Energy Supply per Capita']
    variable.sort_values('Population', ascending=False,inplace =True)
    return variable.iloc[2].name
```

```python
def function_nine():
    variable = function_one()
    variable['Population'] = variable['Energy Supply']/variable['Energy Supply per Capita']
    variable['Citations per person'] = variable['Citations']/variable['Population']
    cor = variable[['Citations per person','Energy Supply per Capita']]
    return cor.corr(method='pearson')
```

```python
function_nine()
```

```python
def function_ten():
    variable = function_one()
    variable['HighRenew'] = variable['%renewable'] > variable['%renewable'].median()
    variable.replace(True,1,inplace=True)
    variable.replace(False,0,inplace=True)
    return variable
```

```python
def function_eleven():
    variable = function_one()
    ContinentDict = {'China':'Asia', 'United States':'North America', 'Japan':'Asia', 'United Kingdom':'Europe',
                     'Russian Federation':'Europe', 'Canada':'North America', 'Germany':'Europe', 'India':'Asia',
                     'France':'Europe', 'South Korea':'Asia', 'Italy':'Europe', 'Spain':'Europe', 'Iran':'Asia',
                     'Australia':'Australia', 'Brazil':'South America'}
    variable['Population'] = variable['Energy Supply']/variable['Energy Supply per Capita']
    groups = pd.DataFrame(columns=['size', 'sum', 'mean', 'std'])
    
    for group, frame in variable.groupby(ContinentDict):
        groups.loc[group] = [len(frame), frame['Population'].sum(), frame['Population'].mean(), frame['Population'].std()]
    
    return groups
```

```python
function_eleven()
```

```python
def function_twelve():
    variable = function_one()
    ContinentDict = {'China':'Asia', 'United States':'North America', 'Japan':'Asia', 'United Kingdom':'Europe',
                     'Russian Federation':'Europe', 'Canada':'North America', 'Germany':'Europe', 'India':'Asia',
                     'France':'Europe', 'South Korea':'Asia', 'Italy':'Europe', 'Spain':'Europe', 'Iran':'Asia',
                     'Australia':'Australia', 'Brazil':'South America'}
    variable['bins'] = pd.cut(variable['%renewable'],5)
    return variable.groupby([ContinentDict,variable['bins']]).size()
```

```python
def function_thirteen():
    variable = function_one()
    ContinentDict = {'China':'Asia', 'United States':'North America', 'Japan':'Asia', 'United Kingdom':'Europe',
                     'Russian Federation':'Europe', 'Canada':'North America', 'Germany':'Europe', 'India':'Asia',
                     'France':'Europe', 'South Korea':'Asia', 'Italy':'Europe', 'Spain':'Europe', 'Iran':'Asia',
                     'Australia':'Australia', 'Brazil':'South America'}
    variable['Population'] = variable['Energy Supply']/variable['Energy Supply per Capita']
    return variable['Population'].apply(lambda x: '{0:,}'.format(x))
```
