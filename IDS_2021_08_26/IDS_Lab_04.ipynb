## Q8

```python
def answer_eight():
    ht_appear = musicians_df.groupby('musician').count()['super_bowl'].reset_index()
    ht_appear = ht_appear.sort_values('super_bowl', ascending=False)
    return ht_appear[ht_appear['super_bowl'] >= 1]
```

```python
answer_eight()
```

## Q9

```python
def answer_nine():
    march_bands = musicians_df[~musicians_df['musician'].str.contains('Marching')]
    march_bands = march_bands[~march_bands['musician'].str.contains('Spirit')]
    
    most_songs = int(max(march_bands['num_songs'].values))
    plt.hist(march_bands['num_songs'].dropna(), bins=most_songs)
    plt.xlabel("Songs Per Show")
    plt.ylabel('Musicians')
    plt.show()
    
    march_bands.sort_values('num_songs', ascending=False, inplace=True)
    display(march_bands.head(15))
```

```python
answer_nine()
```
