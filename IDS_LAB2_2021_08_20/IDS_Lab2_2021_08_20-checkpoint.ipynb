```python
#Q2

def answer_two():
    
    deleted_entries_count =(GDP.shape[0] - final_dataset.shape[0]) + (energy.shape[0] - final_dataset.shape[0])
    return deleted_entries_count

answer_two()
```

```python
#Q3

def answer_three():
    
    final_dataset['avgGDP'] = (final_dataset['2006'] + final_dataset['2007'] + final_dataset['2008'] +final_dataset['2009'] + final_dataset['2010'] + final_dataset['2011'] +final_dataset['2012'] + final_dataset['2013'] + final_dataset['2014'] + final_dataset['2015']) / 10  
    return final_dataset

answer_three()
```

```python
#Q4

def answer_four():
    
    sixth_largest_country_gdp = final_dataset.nlargest(6,'avgGDP').iloc[-1]
    sixth_largest_country_difference = sixth_largest_country_gdp['2015'] - sixth_largest_country_gdp['2006']
    return sixth_largest_country_difference

answer_four()
```

```python
#Q5 

def answer_five():
    
    return final_dataset['Energy Supply per Capita'].mean()

answer_five()
```

```python
#Q6

def answer_six():

    maximum_renewable = final_dataset.nlargest(1,'% Renewable')
    return maximum_renewable['% Renewable']
    
answer_six()
```

```python
# Q7
def answer_seven():
    
    final_dataset['Ratio'] = final_dataset['Self-citations']/final_dataset['Citations']
    maximum_ratio = final_dataset.nlargest(1,'Ratio')
    return maximum_ratio['Ratio']

answer_seven()
```

```python
#Q8

def answer_eight():
    final_dataset['estimated_population'] = final_dataset['Energy Supply'] / final_dataset['Energy Supply per Capita']
    third_populous_country = final_dataset.nlargest(3,'estimated_population').iloc[2:3]
    return tuple(third_populous_country.index)
answer_eight()
```

```python
#Q9

def answer_nine():
    final_dataset['citable document per person'] = final_dataset['estimated_population'] / final_dataset['Citable documents']
    r = final_dataset[['Energy Supply per Capita','citable document per person']]
    return r.corr(method='pearson')

answer_nine()
```

```python
#Q10

def answer_ten():
    
    final_dataset['HighRenew'] = final_dataset['% Renewable'] > final_dataset['% Renewable'].median()
    final_dataset.replace(False,0, inplace=True)
    final_dataset.replace(True,1,inplace=True)
    return final_dataset['HighRenew']

answer_ten()
```
